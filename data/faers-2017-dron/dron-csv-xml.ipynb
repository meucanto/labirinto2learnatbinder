```java
import java.util.ArrayList;
import java.io.PrintWriter;

class Drug {
    public String id;
    public String name;
    public ArrayList<Drug> sub = new ArrayList<Drug>();
    
    public Drug(String id, String name) {
        this.id = id;
        this.name = name;
    }
    
    public void addSub(Drug subDrug) {
        this.sub.add(subDrug);
    }
    
    public void generateXML(PrintWriter saida) {
        System.out.println("<drug id=\"" + this.id + "\" name=\"" + this.name + "\">");
        saida.println("<drug id=\"" + this.id + "\" name=\"" + this.name + "\">");
        sub.forEach((v) -> { 
            v.generateXML(saida);
        });
        System.out.println("</drug>");
        saida.println("</drug>");
    }
}
```

```java
import java.util.Hashtable;
import java.io.BufferedReader;
import java.io.FileReader;
import java.io.IOException;
import java.io.FileWriter;
import java.io.PrintWriter;

try {
    Hashtable<String, Drug> drugs = new Hashtable<String, Drug>();

    FileReader arquivo = new FileReader("drug.csv");
    BufferedReader formatado = new BufferedReader(arquivo);
    String linha = formatado.readLine();
    linha = formatado.readLine();
    while (linha != null) {
        int p = linha.indexOf("\"");
        String fields[];
        if (p == -1)
            fields = linha.split(",");
        else {
            int v = linha.indexOf(",");
            fields = new String[2];
            fields[0] = linha.substring(0, v);
            fields[1] = linha.substring(v+2, linha.length()-1);
        }
        drugs.put(fields[0], new Drug(fields[0], fields[1]));
        linha = formatado.readLine();
    }
    arquivo.close();
    
    /*
    drugs.forEach((k, v) -> { 
       System.out.println(k + "," + v.name);
    });
    */
    
    arquivo = new FileReader("drughierarchy.csv");
    formatado = new BufferedReader(arquivo);
    linha = formatado.readLine();
    linha = formatado.readLine();
    while (linha != null) {
        String fields[] = linha.split(",");
        Drug sub = drugs.get(fields[0]);
        Drug parent = drugs.get(fields[1]);
        parent.addSub(sub);
        linha = formatado.readLine();
    }
    arquivo.close();
    
    Drug root = drugs.get("http://www.w3.org/2002/07/owl#Thing");
    FileWriter arquivoSaida;
    PrintWriter formatadoSaida;
    arquivoSaida = new FileWriter("dron.xml");
    formatadoSaida = new PrintWriter(arquivoSaida);
    root.generateXML(formatadoSaida);
    formatadoSaida.close();
} catch (IOException erro) {
    erro.printStackTrace();
}
```
