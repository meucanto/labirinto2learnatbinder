# Introdução MQTT

MQTT (Message Queuing Telemetry Transport) é um protocolo de mensagens otimizado para redes TCP/IP não confiáveis ou de alta latência. O esquema de troca de mensagens é fundamentado no modelo Publicador-Subscritor, conforme figura abaixo.

![](https://uploads.filipeflop.com/2018/03/MQTT_Diagrama.png)

A comunicação via MQTT é composta das seguintes partes: 

* **publishers:** publicadores de mensagens em tópicos;

* **subscribers:** consumidores interessados em receber mensagens que associadas a tópicos específicos;

* **broker de mensagem:** servidor que gerencia inscrições em tópicos e publicações de mensagens em tópicos.

Mensagens sempre são publicadas em tópicos. Tópicos são divididos em níveis hierárquicos baseados no caracter `\`.

O corpo de uma mensagem que trafega no barramento é conhecido como payload.

## Inicializando um consumidor

O comando **mosquitto_sub -v -t < tópico >** subscreve o terminal ao tópico `<tópico>` e exibe as mensagens que foram enviadas para aquele tópico.


### Exercício:

Abra um terminal e digite os comandos abaixo para iniciar um consumidor interessado em receber as mensagens que estão associadas ao tópico `news/stream`

```bash

mosquitto_sub -v -t "news/stream"


```

**DICA:**  use `shift + insert` para colar os comandos no terminal.

## Inicializando um produtor

O comando **mosquitto_pub -v -t <tópico> -m < payload >** publica uma mensagem com payload `<payload>` no tópico `<tópico>`.
Clientes que assinaram o mesmo tópico receberão esta mensagem.


### Exercício:

Abra um **segundo** terminal e digite os comandos abaixo para iniciar um produtor e publicar uma mensagem no servidor. Observe como o consumidor que está em execução no primeiro terminal receberá a mensagem.
    
 

```bash

mosquitto_pub -t "news/stream" -m "nova notícia"

```







### Incrições múltiplas usando curingas na hirarquia de tópicos

Um consumidor poderá se inscrever em tópicos específicos ou em múltiplos tópicos.
Para inscrever-se em múltiplos tópicos os consumidores podem adotar caracters curingas (wildcards) para se referir a níveis na hierarquia de tópicos. 
Os dois curingas disponíveis são:


* `#` (hash) - multiplos níveis - exemplo:  ` nivel_0/nivel_1/# `
* `+` (mais) - único nível - exemplo:       ` nivel_0/+/nivel_2/nivel_3 `


#### Exercício

Em um terminal, publique todas as mensagens/tópicos abaixo usando `mosquitto_pub` e utilize apenas um consumidor para:

1. receber todas as notícias;
2. receber todas as notícias de futebol (soccer);
3. receber todas as notícias de todos os esportes que o Flamengo possui time.


```bash

mosquitto_pub -t "news/stream/sport/soccer"               -m "nova notícia sobre futebol em geral"
mosquitto_pub -t "news/stream/sport/soccer"               -m "nova notícia sobre futebol em geral 2"
mosquitto_pub -t "news/stream/sport/soccer/vasco"         -m "nova notícia sobre time de futebol do vasco"
mosquitto_pub -t "news/stream/sport/soccer/flamengo"      -m "nova notícia sobre time de futebol do flamengo"

mosquitto_pub -t "news/stream/sport/basketball"           -m "nova notícia sobre basketball em geral"
mosquitto_pub -t "news/stream/sport/basketball/flamengo"  -m "nova notícia sobre time de basquete do flamengo"
mosquitto_pub -t "news/stream/sport/basketball/juventude" -m "nova notícia sobre time de basquete do juventude"
mosquitto_pub -t "news/stream/sport/basketball/vasco"     -m "nova notícia sobre time de basquete do vasco"

mosquitto_pub -t "news/stream/economy"                    -m "nova notícia sobre economia"
mosquitto_pub -t "news/stream/fashion"                    -m "nova notícia sobre moda"

```
